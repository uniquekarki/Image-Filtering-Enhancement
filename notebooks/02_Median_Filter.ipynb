```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../images/img2.jpg')
```

```python
filter_image = cv2.medianBlur(image, 5)
```

```python
concat_image = cv2.hconcat([image, filter_image])
```

```python
plt.imshow(concat_image)
plt.title('Original VS. Median Filter Image')
plt.show()
```
