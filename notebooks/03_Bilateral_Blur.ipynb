```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/taj.jpg')
```

```python
bilateral_image = cv2.bilateralFilter(image, 15, 75, 75) 
```

```python
concat_image = cv2.hconcat([image, bilateral_image])
```

```python
plt.imshow(concat_image)
plt.title('Original Vs. Bilateral Blur')
plt.show()
```
